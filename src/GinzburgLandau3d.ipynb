# The Ginzburg-Landau model of a classical second-order phase transition ($d=3$ spatial dimensions) #

Introductory text below is the same as the $d=1$ example.

A phase transition is a point in thermodynamic-parameter space where the log marginal likelihood ($=-F/T$, where $F$ is the free energy and $T$ is the temperature) is a non-analytic function of one of its thermodynamic parameters. Here, a thermodynamic parameter is a macroscopic quantity such as temperature $T$, pressure $P$, the (expected) global order parameter $\Psi$, etc. The global order parameter is the spatial mean of the local order parameter, which is defined at each lattice point and would be something like the local magnetization (or spin). The local order parameter on an $N$-element lattice is what a statistician would refer to as the $N$-dimensional parameter of the model; the lattice and order parameters will become clear below.

Phase transitions are either quantum or classical. Quantum phase transitions occur at $T=0$ and are driven by quantum fluctuations, which are a consequence of the Heisenberg uncertainty principle. Classical phase transition occur at $T> 0$ and are driven by thermal fluctuations, which is heat energy transferred from the system's surroundings.

Classical phase transitions are either classified as continuous or non-continuous. At a non-continuous phase transition, the structure of the material changes: e.g., at the solid-liquid water transition, the structure of the material changes from a rigid lattice (strong bonds between the molecules) at low temperature to a liquid (weak bonds between the molecules) at high temperature. Breaking the strong bonds to enter the liquid phase requires energy, so transferring energy to the system at the transition temperature doesn't increase the temperature of the system until the liquid phase has been entered (temperature is the expected kinetic energy of the particles). This energy transfer at constant temperature is called latent heat. A classical phase transition with non-zero latent heat is non-continuous; one with zero latent heat is continuous. Below, we consider continuous classical phase transitions.

A continuous phase transition is classified by its order ('order' will become clear later on). The Ginzburg-Landau model of phase transitions is a phenomenological model, which means it's constructed from the top down to reproduce the physics we're interested in: in this case, that of a many-body system at and around a second-order phase transition. (The model can be generalized to higher-order phase transitions, but can't describe topological phase transitions such as the Berezinskii-Kosterlitz-Thouless transition, which are infinite order.) The model averages the local order parameter over mesoscopic volumes $a^d$ to generate a lattice model of the system, where $a\in \mathbb{R}_{>0}$ is the lattice spacing and $d\in \mathbb{N}$ is the dimension of the system's position space. Systems that share the same dimension in both their position space and their order-parameter space (each of which do not have to have the same dimension within one system) are in the same universality class and therefore share the same mesoscopic physics (up to the values of the coefficients $\alpha$, $\lambda$ and $\tau$ introduced below). Ginzburg-Landau theory was designed to be a general mesoscopic model of phase transitions for systems in the same universality class.

We define $\Omega \subset a\mathbb{Z}^d$ to be the set of all lattice points, $\psi : \Omega \rightarrow \mathbb{R}$ to be the local order parameter of the model, and use $x\in \Omega$ to denote the lattice points. The global order parameter is then $\Psi:=\sum_{x\in \Omega}\psi(x)/N$, where $N$ is the number of lattice sites. Note that the local order parameter can in general be a lattice vector field but a scalar field produces a double well at each lattice site in the low-temperature phase - rather than a 'Mexican hat' - and so should generate a more challenging marginal 
distribution. We then define our potential in non-zero external field $h\in \mathbb{R}$:

\begin{align}
U_h[\psi (x)]=\beta a^d \sum_{x\in \Omega} \left[ \frac{\tau}{2} \left[ \psi(x) \right]^2+ \frac{\alpha}{2} \| \tilde{\nabla} \psi(x) \|^2 +\frac{\lambda}{4} \left[ \psi(x) \right]^4 - h\psi(x) \right],
\end{align}

where $\tilde{\nabla}$ is the forward finite-difference operator, $\beta := 1/T$ is the inverse temperature, $T\in \mathbb{R}_{>0}$ is the temperature, $\tau := (T-T_{\rm c})/T_{\rm c}$ is the reduced temperature and $\alpha , \lambda \in \mathbb{R}_{>0}$ are the coefficients of the gradient and quartic terms: each are strictly positive so that our marginal distribution is well defined (quartic term) and to induce correlations (the gradient term). Here, the reduced temperature is reduced with respect to the transition temperature $T_{\rm c} \in \mathbb{R}_{>0}$. The quadratic term then induces order/disorder in the low/high-temperature phase. N.B.: we impose periodic boundary conditions on the lattice and the local order parameter: $\psi(0,0,0)$ is a nearest neighbour of $\psi(L-a,0,0)$ in $d=3$ spatial dimensions.

The external field $h$ represents a real physical field (e.g., a global external magnetic field that couples to each microscopic magnetic moment / spin in a magnetic system) but is also the generator for the moment generating function once the likelihood has been constructed. The marginal likelihood for $h\ne 0$ is

\begin{align}
Z_h = \int \mathcal{D} \psi \exp \{ -U_h[\psi (x)] \},
\end{align}

where $\int \mathcal{D}\psi :=\prod_{x\in \Omega}\left[ \int_{\mathbb{R}}d\psi(x) \right]$. Derivatives of the free energy $F_h:=-\beta^{-1}\ln Z_h$ with respect to the external field $h$ then generate different centred moments:

\begin{align}
\partial_h F_h = N \mathbb{E}_h \Psi ,
\end{align}

and

\begin{align}
\partial_h^2 F_h = -N^2 \beta Var_h \Psi ,
\end{align}

where the subscript $h$ denotes expected values with $h\ne 0$.

We define the susceptibility $\chi (h,T):= \beta Var_h \Psi$ so that

\begin{align}
\chi (h,T) = -\frac{1}{N} \partial_h \mathbb{E}_h \Psi .
\end{align}

Now, at $T<T_{\rm c}$ the system is ordered, and hence $\lim_{h\rightarrow 0^{\pm}} \left[\mathbb{E}_h \Psi \right] = \pm c$ ($c\in \mathbb{R}_{>0}$); at $T>T_{\rm c}$ the system is disordered, and hence $\lim_{h\rightarrow 0}\left[\mathbb{E}_h \Psi \right] = 0$. (N.B.: $\mathbb{E} \Psi  = 0$ at $T<T_{\rm c}$, where $\mathbb{E}$ denotes the expectation at $h=0$. This is since the potential is symmetric with respect to $\psi (x) \leftrightarrow -\psi (x) \, \forall x\in \Omega$; however, the expected value is non-zero on experimental timescales because it's hard for the real system to hop between the double wells: this 'spontaneous symmetry breaking' breaks ergodicity on experimental timescales in the low-temperature phase.) It follows that

\begin{align}
\lim_{T \rightarrow T_{\rm c}^{-}} \left[ \lim_{h\rightarrow 0^{\pm}} \left[ \chi (h,T< T_{\rm c}) \right] \right] = \mp \infty ,
\end{align}

and we've finally arrived at the punchline: the second derivative of the log marginal likelihood with respect to the external field $h$ diverges at the transition (where $h=0$ and $T=T_{\rm c}$) hence we have a second-order phase transition. Referring back to the first paragraph, the phase transition is the point in thermodynamic-parameter space at which the Taylor expansion of the log marginal likelihood with respect to the expectation of the global order parameter $\mathbb{E} \Psi$ breaks down because the second-order coefficent diverges.

For our investigations, we will work in $d=3$ spatial dimensions, set $a=1$ and $h=0$, and define $U[\psi (x)]$, $Z$ and $F$ to be the potential, marginal likelihood and free energy in zero external field, respectively. We set $T_{\rm c} =1$ so that the potential becomes

\begin{align}
U[\psi (x)]=& \sum_{x\in \Omega} \left[ \frac{1}{2} (1-\beta) \left[ \psi(x) \right]^2+ \frac{1}{2} \beta \alpha \| \tilde{\nabla} \psi(x) \|^2 +\frac{1}{4}\beta \lambda \left[ \psi(x) \right]^4 \right] .
\end{align}

We now compute the gradient vector:

\begin{align}
\frac{\delta U [\psi (y)]}{\delta \psi (x)} = (1-\beta)\psi(x) - \beta \alpha \nabla^2 \psi(x) +  \beta \lambda \left[ \psi(x)\right]^3 ,
\end{align}

where $\delta / \delta \psi (x)$ denotes a functional derivative with respect to the local order parameter at site $x\in \Omega$, so we have $N$ auxiliary momenta. Here, $\nabla^2:=\hat{\nabla}\cdot \tilde{\nabla}$, where $\hat{\nabla}$ is the backwards finite-difference operator [$\nabla^2 f(x) = f(x+a)+f(x-a) -2f(x)$ for some lattice scalar field $f: \Omega \rightarrow \mathbb{R}$ on a $(d=1)$-dimensional lattice].

In the low-temperature phase, the marginal distribution of each local order parameter is double well, each of which is coupled to its (nearest) neighbour local order parameters by the gradient term. We plan to simulate the low-temperature phase with coefficient $\lambda$ such that each hill between each of the double wells is tall and narrow in order to generate a challenging target distribution. This will demonstrate that the algorithm can still perform well in the centre of the space of a challenging target distribution. We also plan to simulate the high-temperature phase while switching between the ideal kinetic energy for Gaussian and quartic marginal distributions. This will allow us to make a comment on kinetic-energy switching to avoid repeated instabilities in the centre of the space (this is unavoidable in the low-temperature phase, but the switching is more to show that we've thought about everything, where possible). 

This version simulates $d=3$ dimensions.

# Create potential class for d=3 array of marginal parameters with periodic boundaries #

```python
# Create potential class for the d=3 Ginzburg-Landau model of a second-order phase transition (1D order paramter);
class GL3d:
    """d=3 Ginzburg-Landau model with 1D order parameter"""
    def __init__(self, lamb, beta, alpha, Len):
        self.lamb = lamb
        self.beta = beta
        self.alpha = alpha
        # Lattice length 'Len' ==> Len x Len x Len lattice points for d=3 (we set a=1)
        self.Len = Len
    
    # Periodic boundaries for our Len^3 length 1D array
    def __pos_x(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len)) # reshapes to a Len*Len*Len matrix
        b = np.pad(a,(0,1),mode='wrap') # copies the 0th entry at each matrix level to (Len+1)th entry
        c = np.delete(b,Len,0) # deletes the (Len)th entry at the highest matrix level
        d = np.delete(c,Len,1) # deletes the (Len)th entry at the second-highest matrix level
        e = np.delete(d,0,2) # deletes the 0th entry at the lowest matrix level
        return np.reshape(e,Len**3) # reshapes to an Len**3-dim vector
    
    def __pos_y(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len))
        b = np.pad(a,(0,1),mode='wrap')
        c = np.delete(b,Len,0)
        d = np.delete(c,0,1)
        e = np.delete(d,Len,2)
        return np.reshape(e,Len**3)

    def __pos_z(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len))
        b = np.pad(a,(0,1),mode='wrap')
        c = np.delete(b,0,0)
        d = np.delete(c,Len,1)
        e = np.delete(d,Len,2)
        return np.reshape(e,Len**3)
    
    def __neg_x(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len))
        b = np.pad(a,(1,0),mode='wrap')
        c = np.delete(b,0,0)
        d = np.delete(c,0,1)
        e = np.delete(d,Len,2)
        return np.reshape(e,Len**3)
            
    def negy(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len))
        b = np.pad(a,(1,0),mode='wrap')
        c = np.delete(b,0,0)
        d = np.delete(c,Len,1)
        e = np.delete(d,0,2)
        return np.reshape(e,Len**3)
    
    def __neg_z(self,psi):
        Len = self.Len
        a = np.reshape(psi,(Len,Len,Len))
        b = np.pad(a,(1,0),mode='wrap')
        c = np.delete(b,Len,0)
        d = np.delete(c,0,1)
        e = np.delete(d,0,2)
        return np.reshape(e,Len**3)

    # Potential function
    def value(self, psi):
        lamb = self.lamb
        beta = self.beta
        alpha = self.alpha
        Len = self.Len
        z1 = (1 - beta)
        z2 = beta * alpha
        z3 = beta * lamb
        psi_posx = self.__pos_x(psi)
        psi_posy = self.__pos_y(psi)
        psi_posz = self.__pos_z(psi)
        vec = np.empty(Len**3)
        vec = 0.5 * z1 * psi**2 + 0.5 * z2 * ((psi_posx - psi)**2 + (psi_posy - psi)**2 + (psi_posz - psi)**2) + 0.25 * z3 * psi**4
        return sum(vec)

    # Gradient function
    def grad(self, psi):
        lamb = self.lamb
        beta = self.beta
        alpha = self.alpha
        Len = self.Len
        z1 = (1 - beta)
        z2 = beta * alpha
        z3 = beta * lamb
        psi_posx = self.__pos_x(psi)
        psi_posy = self.__pos_y(psi)
        psi_posz = self.__pos_z(psi)
        psi_negx = self.__neg_x(psi)
        psi_negy = self.negy(psi)
        psi_negz = self.__neg_z(psi)
        vec = np.empty(Len**3)
        vec = z1 * psi - z2 * (psi_posx + psi_negx + psi_posy + psi_negy + psi_posz + psi_negz - 6 * psi) + z3 * psi**3
        return vec
    
beta = 1.5 # (1-beta)=-0.5; the low-temperature phase, but without too steep a hill between each marginal double well.
alpha = 1.0 # An order of magnitude smaller than the marginal coefficients.
lamb = 1.0 # The same order of magnitude as the quadratic coefficient.
Len = 10 # 10x10x10 3D lattice
U = GL3d(lamb, beta, alpha, Len)
# array = 100.0 * np.ones(Len**3)
# print U.value(array)
# print U.grad(array)
# print array[Len**3-1]
```

## Average ESS at equilibrium, averaged over 10 runs.

We run 10 chains for each kinetic energy, using the optimal step size, compute the minimum, mean and maximum ESS for each chain, store them all and have a look at the results.

```python
import MC_diag as d

# Set up experiment
chains = 10
Len = 10
start_vec = np.zeros(Len**3)
niter = 10000

# Potential class
beta = 2.0 # (1-beta)=-1.0; the low-temperature phase
alpha = 0.1 # An order of magnitude smaller than the marginal coefficients.
lamb = 0.5 # The same order of magnitude as the quadratic coefficient.
Len = 10 # 10x10x10 3D lattice
U = GL3d(lamb, beta, alpha, Len)

# Kinetic energies
K_g = Gauss()
K_PowRel = GR_family(4/3.0)
K_PowExp = PE_family(4/3.0)
K_Rel = GR_family(1.0)
```

#### Gaussian kinetic energy

```python
# ESS loop for Gaussian
ESS_g = np.empty([chains,Len**3])
ESS_summary_g = np.empty([chains,3])
stepsize = 0.20

for i in xrange(chains):
    print "Iteration: %d" % i
    samples = h.HMC(niter, start_vec, U, K_g, L = 10, e = stepsize)
    x_samples = samples['x']
    for j in xrange(Len**3): # may be a more efficient way to do this
        ESS_g[i,j] = float(np.asarray(d.ess(x_samples[j])))
    ESS_summary_g[i,:] = [np.min(ESS_g[i,:]),np.mean(ESS_g[i,:]),np.max(ESS_g[i,:])]
    print "Effective sample summaries: %d, %d, %d" % (ESS_summary_g[i,0],ESS_summary_g[i,1],ESS_summary_g[i,2])

np.mean(ESS_summary_g,axis = 0)
```
